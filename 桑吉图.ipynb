```python
import pandas as pd
from pandas import DataFrame as df
```

```python
de = pd.read_excel('D:/Users/pjchang/Downloads/test.xlsx')
##df = de[de["num"]>10]
df = de[["area","countryname","num"]]
df
```

```python
nodes = []
for i in range(2):
    values = df.iloc[:,i].unique()
    for value in values:
        dic = {}
        dic["name"]=value
        nodes.append(dic)
nodes
```

```python
linkes = []

for i in df.values:
    dic = {}
    dic['source'] = i[0]
    dic['target'] = i[1]
    dic['value'] = i[2]
    linkes.append(dic)
linkes
```

```python
from pyecharts.charts import Sankey
from pyecharts import options as opts

pic = (
Sankey().add
    ("Sankey",
    nodes,
    linkes,
    ##linestyle_opt = opts.LineStyleOpts(opacity = 0.3, curve = 0.8, color = 'source'),
    label_opts=opts.LabelOpts(position='right'),
    ##node_gap = 30,
    )
    .set_global_opts(title_opts=opts.TitleOpts(title = '语言线_全球旅客评价流向'))
    .render('语言线_全球旅客评价流向1.html')    
)
```
